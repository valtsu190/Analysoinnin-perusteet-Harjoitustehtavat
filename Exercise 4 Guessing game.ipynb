```python
print("Guess a name")

name = "Juho"
arvaus = ""
counter = 0

while(arvaus != name):
    arvaus = input("Guess a name")
    if(arvaus != name):
        i = input("Do you want to quit (y/n)?")
        if (i == "y"):
            break
    counter = counter + 1

if(arvaus == name):
    print("Correct! You guessed:", counter, "times.")

```
